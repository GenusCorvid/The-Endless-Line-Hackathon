```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
final_data_set = pd.read_csv('final_data/merged_datasets_with_events.csv',sep = ',',encoding='utf-8')
final_data_set.columns
```

```python
final_data_set.sort_values(by='DEB_TIME', ascending=True, inplace=True)
final_data_set.head(50)
```

```python
# Aggregate to daily attendance (e.g., summing the 15-minute slices)
df_daily = final_data_set.groupby('WORK_DATE')['ATTENDANCE'].first().reset_index()

# Create a full date range covering the whole period
full_dates = pd.date_range(start=df_daily['WORK_DATE'].min(),
                           end=df_daily['WORK_DATE'].max(),
                           freq='D')
df_full = pd.DataFrame({'WORK_DATE': full_dates})

df_daily['WORK_DATE'] = pd.to_datetime(df_daily['WORK_DATE'])
df_full['WORK_DATE'] = pd.to_datetime(df_full['WORK_DATE'])

# Merge to include all days; days missing in df_daily will have NaN attendance
df_full = df_full.merge(df_daily, on='WORK_DATE', how='left')

# Fill missing attendance with 0
df_full['ATTENDANCE'] = df_full['ATTENDANCE'].fillna(0)

# Define the COVID start date in France.
covid_start = pd.to_datetime("2020-01-24")

# Create a dummy variable: 0 before COVID, 1 on/after COVID
df_full['covid'] = (df_full['WORK_DATE'] >= covid_start).astype(int)

# Create the "days since COVID" variable. For dates before COVID, it remains 0.
df_full['days_since_covid'] = (df_full['WORK_DATE'] - covid_start).dt.days.clip(lower=0)

# Create a set of dates that were originally present in your data
original_dates = set(df_daily['WORK_DATE'])

# For dates after COVID began, flag as lockdown if the day is not in the original data.
df_full['lockdown'] = 0
df_full.loc[
    (df_full['WORK_DATE'] >= covid_start) & (~df_full['WORK_DATE'].isin(original_dates)),
    'lockdown'
] = 1

df_full['lockdown_interaction'] = df_full['lockdown'] * df_full['days_since_covid']

#df_full['days_since_covid2'] = df_full['days_since_covid']**2

# Rename for Prophet
df_full.rename(columns={'WORK_DATE': 'ds', 'ATTENDANCE': 'y'}, inplace=True)

```

```python
#!py -3.10 -m pip install prophet
from prophet import Prophet

# Initialize the Prophet model
m = Prophet()

# Add extra regressors
m.add_regressor('covid', mode='multiplicative')
m.add_regressor('days_since_covid', mode='multiplicative')
m.add_regressor('lockdown', mode='multiplicative')
#m.add_regressor('lockdown_interaction', mode='multiplicative')
#m.add_regressor('days_since_covid2')

# Fit the model on the full dataset
m.fit(df_full)
```

```python
# Create a future DataFrame
future = m.make_future_dataframe(periods=1095)

# Set the extra regressors for the forecast period
future['covid'] = (future['ds'] >= covid_start).astype(int)
future['days_since_covid'] = (future['ds'] - covid_start).dt.days.clip(lower=0)
#future['days_since_covid2'] = future['days_since_covid']**2

# Assume no lockdowns in the forecast period; adjust as needed.
future['lockdown'] = 0
future['lockdown_interaction'] = future['lockdown'] * future['days_since_covid']

# Make predictions
forecast = m.predict(future)

# Plot the forecast
fig1 = m.plot(forecast)
#fig = m.plot_components(forecast)

forecast.to_csv('final_data/attendanceforecast.csv', index=False)
```

```python
#Train test split (Not very useful as post covid data is very limited)
# Assume df_full is your full dataset with the engineered regressors and renamed columns.
train = df_full[df_full['ds'] < '2021-08-01']  # for example, train on data before July 2021
test = df_full[df_full['ds'] >= '2021-08-01']
# Fit the model on the training data
m = Prophet()
m.add_regressor('covid', mode='multiplicative')
m.add_regressor('days_since_covid', mode='multiplicative')
m.add_regressor('lockdown', mode='multiplicative')
#m.add_regressor('lockdown_interaction', mode='multiplicative')
m.fit(train)

# Create future dates covering the test period
future = test[['ds']].copy()
future['covid'] = (future['ds'] >= pd.to_datetime("2020-01-24")).astype(int)
future['days_since_covid'] = (future['ds'] - pd.to_datetime("2020-01-24")).dt.days.clip(lower=0)
# Here, you should set 'lockdown' appropriately for your test period.
future['lockdown'] = 0  # or the known value if a lockdown was in effect
future['lockdown_interaction'] = future['lockdown'] * future['days_since_covid']

# Generate forecasts
forecast = m.predict(future)

# Compare forecast 'yhat' to the actual 'y' in your test set:
test = test.reset_index(drop=True)
forecast = forecast.reset_index(drop=True)
results = test[['ds', 'y']].copy()
results['yhat'] = forecast['yhat']
#print(test)
results.to_csv('final_data/attendencetestresults.csv', index=False)
```
